# Function for conversion of Strings to num

```python
    def convert(s):
    #try:
        ss = s.split(' ')
        a = list(ss)
        
        str1 = ""
        for j in a:
            for i in j:
                if (i.isdigit() or i == "."):
                    str1 += i
        return float(str1)
    #except:
     #   return None
```

# Testing my function

# Changing String values of mileage,engine,power,seats,kmdriven,selling price to Float

```python
df['n_mileage'] = df['mileage'].apply(convert)
```

```python
df['n_engine'] = df['engine'].apply(convert)
```

```python
df['n_power'] = df['max_power'].apply(convert)
```

```python
df['n_seats'] = df['seats'].apply(convert)
```

```python
df['n_km_driven'] = df['km_driven'].apply(convert)
df['n_selling_price'] = df['selling_price'].apply(convert)
```

```python
df1 = df.drop(['mileage','engine','max_power','seats','new_price','km_driven','selling_price'],axis ='columns')
```

```python
df1
```

# LabelEncoding Normal Data

```python
from sklearn.preprocessing import LabelEncoder
le_transmission = LabelEncoder()
le_seller = LabelEncoder()
le_fuel = LabelEncoder()
le_owner = LabelEncoder()
df1['transmission'] = le_transmission.fit_transform(df1['transmission_type'])
df1['seller'] = le_seller.fit_transform(df1['seller_type'])
df1['fuel'] = le_fuel.fit_transform(df1['fuel_type'])
df1['owner_type'] = le_fuel.fit_transform(df1['owner_type'])
```

```python
df2 = df1.drop(['seller_type','fuel_type','transmission_type'],axis = 'columns')
df2
```

# Sorting Name Strings and reducing them

```python
name_stat = df2.groupby('full_name')['full_name'].agg('count')
name_stat<=2
```

```python
def name(s):
    ss = s.split(' ')
    return (ss[0]+" "+ss[1])
```

# OneHotEncoding

```python
dummies = pd.get_dummies(df2.full_name)
```

```python
dummies
```

```python
df3 = pd.concat([df2,dummies],axis='columns')
```

```python
df3
```

```python
df4 = df3.drop(['full_name'],axis='columns')
df4
```

# Converting price to Lakhs

```python
df4['n_selling_price'] = df4['n_selling_price'].apply(lambda x : x*100000)
```

# Adding New Column for KM/Price

```python
df4['kmperprice'] = df4['n_km_driven']/df4['n_selling_price']
df4
```

# Removing datas that are causing confusion to our model

```python
df4 = df4[df4.kmperprice<1]
df4
```

```python
target = df4['n_selling_price']
```

```python
df5 = df4.drop(['n_selling_price'],axis='columns')
```

```python
df5
```

```python
def convert_year(i):
    return (2021-i)
```

```python
df5['age'] = df5['year'].apply(convert_year)
df6 = df5.drop(['year','owner_type'],axis = 'columns')
df6
```

```python
df6.describe()
```

```python
df6
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df6,target,test_size=0.2,random_state=42)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor 
GradientBoostingRegressor()
model = GradientBoostingRegressor(n_estimators=80)
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, df6, target, cv=10)
scores.mean()
```
